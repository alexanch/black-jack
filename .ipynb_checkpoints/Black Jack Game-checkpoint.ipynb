```python
import random
from IPython.display import clear_output
import glob, os
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
import pandas as pd
import sys
from io import StringIO
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

suits = ('Spades','Clubs','Diamonds', 'Hearts')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

# os.chdir("c:/Users/Alexanch/Desktop/proj/black-jack/cards")
# loc = "c:/Users/Alexanch/Desktop/proj/black-jack/cards/"
# lst = glob.glob('*.jpg')
# lst2 = [loc+x for x in lst]
```

```python
class Card:

    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank + ' of ' + self.suit
        #return self.link

class Deck:
    
    def __init__(self):
        self.deck = []  # start with an empty list
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
                
    def __str__(self):
        deck_comp = ''  # start with an empty string
        for card in self.deck:
            deck_comp += '\n '+card.__str__() # add each Card object's print string
        return 'The deck has:' + deck_comp

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        temp = self.deck.pop()
        return temp

class Hand:
    
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.score = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    
    def add_card(self,card):
        self.cards.append(card)
        self.score += values[card.rank]
        if card.rank == 'Ace':
            self.aces += 1  # add to self.aces
    
    
    
    def adjust_for_ace(self):
        while self.score >21 and self.aces:
            self.score -= 10
            self.aces -= 1    


class Chips:
    
    def __init__(self, value = 100):
        self.total = value  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input('Enter your bet:'))
        except:
            print('This is not a number.')
        else:
            if chips.bet > chips.total:
                print ('The balance is lower than the bet.')
            else:
                break

# def plot_cards():
    
#     fig = plt.figure(figsize=(32, 32))

#     for i in range(1,len()):
#                 fig.add_subplot(2, 4, i)
#                 plt.imshow(plt.imread(lst2[i]))
#                 plt.xticks([])
#                 plt.yticks([])
#                 plt.tight_layout()
                
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()

def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop
    while True:
        d = input('Hit or stand?').lower()[0]
        if d == 'h':
            hit(deck,hand)
        elif d == 's':
            print("Dealer's turn.")
            playing = False
        else:
            print("Sorry, please try again.")
            continue
        break

        
def show_some(player,dealer):
    print("\nPlayer's Hand:", *player.cards, sep='\n ')
    print("\nDealer's Hand:\n <card hidden>\n",dealer.cards[1])  
    
def show_all(player,dealer):
    print("\nPlayer's cards:",*player.cards, sep='\n')
    print("\nDealer's cards:",*dealer.cards, sep='\n')

def player_busts(chips):
    print('You lost!')
    chips.lose_bet()

def player_wins(chips):
    print('You won!')
    chips.win_bet()

def dealer_busts(chips):
    print('Dealer lost!')
    chips.win_bet()
    
def dealer_wins(chips):
    print('Dealer won!')
    chips.lose_bet()
    
def push():
    print("It's a Tie!")
```

```python
# Set up the Player's chips
player_chips = Chips(value = 1000)

# Print an opening statement
name = input('What is your name?')
clear_output()
print(f"Hello, {name}")

# play game until player stops  
while True:
    # Create & shuffle the deck
    deck = Deck()
    deck.shuffle()
    # Deal two cards to player and pc
    player = Hand()
    player.add_card(deck.deal()) 
    player.add_card(deck.deal())
    
    pc = Hand()
    pc.add_card(deck.deal()) 
    pc.add_card(deck.deal())
    
    # Take a bet from the player
    print ('\nYour balance:',player_chips.total)
    take_bet(player_chips)
    
    # Show cards, one dealer's card is hidden
    show_some(player,pc)
    
    # hit until stand is chosen. 
    playing = True
    while playing:
        
        # Hit or Stand
        hit_or_stand(deck,player)
        # Show cards
        clear_output()
        show_some(player,pc)
        # If player's hand exceeds 21, finish the game - player lost
        if player.score>21:
            clear_output() 
            player_busts(player_chips)
            break
            
       
    # pc(dealer)'s turn
    if player.score<=21:
        clear_output() 
        #hit until total score is lower than 18
        while pc.score<18:
            hit(deck,pc)
        # lost if score is > 21
        if pc.score > 21:
            dealer_busts(player_chips)
       
        # compare player's and pc's scores
        elif pc.score < player.score:
            player_wins(player_chips)
        elif pc.score > player.score:
            dealer_wins(player_chips)
        else:
            #if equal:
            push()
    
    #show all cards and balance
    show_all(player,pc)
    print ('\nYour balance:',player_chips.total)

    # play again?
    temp = input(f'Another game, {name}? "Y" to continue.')
    if temp[0].lower() == 'y':
        playing = True
        clear_output()
        continue
    else:
        clear_output()
        print(f'Have a nice day, {name}!')
        break
        
```
